# Actividad

El objetivo de esta actividad es obtener el resultado de las siguientes métricas:
## Query 1:
Obtener el número de taxis distintos que han hecho un viaje de entre 20 y 44 € y hayan pagado con “Credit Card” (pista: utilizar el valor taxi_id, en la plataforma de pago tened en cuenta las mayúsculas y las minúsculas).


## Query2:
Obtener el número de taxis distintos que han hecho un viaje que hayan pagado en extras menos de 10€ y más de 60€ y el tipo de pago sea “Cash” (pista: utiliza el valor taxi_id, en el método de pago tened en cuenta las mayúsculas y las minúsculas)
El campo unique_key representa a un identificador único que se le asigna a cada uno de los viajes realizados.

```python
# importar librerias Pandas
import pandas as pd
```

```python
file_query1 = pd.read_csv('actividad-clase-2.csv',sep=',', encoding = "ISO-8859-1", header=0, low_memory=False)
file_query2 = pd.read_csv('actividad-clase-2.csv',sep=',', encoding = "ISO-8859-1", header=0, low_memory=False)
```

## Query 1
trip_total entre 20 y 44 -- Rango
and payment_type = Credit Card
nota: taxi_id valores unicos

```python
# Crear dataframe a partir del filtros
df_query1 = pd.DataFrame(file_query1[(file_query1['trip_total'] > 20) & (file_query1['trip_total'] < 44) & (file_query1['payment_type'].str.contains('Credit Card')) & (file_query1['payment_type'].str.contains('Credit Card'))])
df_query1
```

```python
df_query1_resultado = pd.DataFrame(df_query1.drop_duplicates(subset = 'taxi_id'))
df_query1_resultado
```

## Resultado Query 1

```python
print('Cantidad de taxis: ', df_query1_resultado['unique_key'].count())
```

## Query 2
extras menor a 10 and mayores 60 -- Rango
<br> Y payment_type = Cash
<br> Nota: taxi_id valores unicos
<br> <b>Comentario</b>: No hay valores de extras mayores a 60 en la fuente de datos.

```python
df_query2 = pd.DataFrame(file_query2[(file_query2['extras'] < 10) & (file_query2['payment_type'].str.contains('Cash'))])
df_query2
```

```python
df_query2_resultado = pd.DataFrame(df_query2.drop_duplicates(subset = 'taxi_id'))
df_query2_resultado
```

## Resultado Query 2

```python
print('Cantidad de taxis: ', df_query2_resultado['unique_key'].count())
```
